```python
import pandas as pd
```

```python
df = pd.read_excel("ESD.xlsx")
df.head()
```

```python
# Make new column for Bonus that they recieve Bonus or not
df.loc[(df["Bonus %"] == 0), "GetBonus"] = "No"
df.loc[(df["Bonus %"] > 0), "GetBonus"] = "Yes"
print(df.head())
```

```python
# Make Series -> Make dataframe -> apply map function to each value
data = {"Months":["January","February","March"]}

df = pd.DataFrame(data)

# Map function
def extract(value):
    return value[:3]

df["short_month"] = df["Months"].map(extract)
df
```
