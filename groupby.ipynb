```python
import pandas as pd 
```

```python
df = pd.read_excel("ESD.xlsx")
df.head()
```

```python
# Get name of columns
df.columns
```

```python
# Count no. of employees per department
gp = df.groupby("Department").agg({"EEID":"count"})
gp
```

```python
# No. of employees gender wise in each department
gp = df.groupby(["Department","Gender"]).agg({"EEID":"count"})
gp
```

```python
# Average age country wise
gp1 = df.groupby("Country").agg({"Age":"mean"})
gp1
```

```python
# Maximum salary per country wise 
gp2 = df.groupby("Country").agg({"Annual Salary":"max"})
gp2
```

```python
# Mean salary of each country's gender
gp3 = df.groupby(["Country","Gender"]).agg({"Annual Salary":"mean"})
gp3
```

```python
# Maximum salary with minimum age of each employee country and gender wise 
gp4 = df.groupby(["Country","Gender"]).agg({"Annual Salary":"max","Age":"min"})
gp4
```
