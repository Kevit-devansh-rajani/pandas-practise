```python
import matplotlib.pyplot as plt
```

```python
# Line Chart: to show trends over time

x = [1, 2, 3, 4, 5]
y = [100, 120, 110, 150, 170]

plt.figure(figsize=(3,3))
plt.plot(x, y,  marker='o')
plt.xlabel('Time')
plt.ylabel('Value')
plt.title("Sales over time")
plt.grid(False)
plt.show()
```

```python
# Bar chart: to compare values over categories
categories = ['A', 'B', 'C']
values = [10, 25, 15]

plt.figure(figsize=(3,3))
plt.title("Category Comparison")
plt.xlabel("Category")
plt.ylabel("Value")
plt.bar(categories, values, color='skyblue')
plt.show()
```

```python
# Histogram: understand distribution of numerical data
import numpy as np

data = np.random.normal(50, 10, 200)

plt.figure(figsize=(3,3))
plt.hist(data, bins=20, color='orange', edgecolor='black')
plt.title("Data Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# Scatter plot: to observe relationship between two variables
x = [1, 2, 3, 4, 5]
y = [2, 4, 5, 4, 5]

plt.figure(figsize=(3,3))
plt.scatter(x, y, color='orange')
plt.title("Scatter Plot Example")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
